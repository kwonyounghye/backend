```python
t1=(1,2,3)
n1,n2,n3=t1 #1개의 튜플 변수로 저장됨
print(f'{n1} {n2} {n3}') #format string
n1, n2=n2, n1
print(f'{n1} {n2}')
```

```python
#{key:value}
a={'name':'김철수'}
print(a)
print(type(a)) #자료형
a['age']=20 #딕셔너리[key]
print(a)
```

```python
a={'name':'김철수', 'phone':'01000000000', 'birth':'990721'}
print(a['name']) #딕셔너리[key]
print(a.get('name')) #딕셔너리.get(key)
print(a.values()) #value 리스트
print(a.keys()) #key 리스트
print(a.items()) #아이템(key-value) 리스트
```

```python
import time #모듈 참조
start = time.time() #현재 시각의 타임스탬프
print(start)
for a in range(1000): #0~999
    print(a, end='  ') #end 줄바꿈 대체 문자

end=time.time()
print() #줄바꿈
print(end-start)
```

```python
print('1')
time.sleep(1) #1초간 멈춤
print("2")
time.sleep(5)
print('7')
```

```python
import time
#range(start, stop, step)
for dan in range(2,10): #2부터 9까지
    print(dan,'단')
    for i in range(1, 10): #1부터 9까지
        print(dan, '*', i, '=', dan*i)
        time.sleep(0.2) #0.2초 멈춤
    print() #줄바꿈
    time.sleep(1)
```

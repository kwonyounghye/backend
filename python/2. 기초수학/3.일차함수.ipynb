```python
a=[1,2,3,4,5] #zip함수 :작은 쪽을 기준으로 하고 나머지는 무시
b=['a','b','c']
c=['딸기', '복숭아', '참외']
list(zip(a,b,c)) #자료를 만들려면 list안에 넣어야 함
```

```python
def my_zip(a,b):
    if len(a)!=len(b):
        print('두 입력값의 길이가 다릅니다.')
    else:
        n=len(a) #자료수
        c=[]
        for i in range(n):
            c.append((a[i],b[i])) #리스트에 튜플 추가
        return c
a=[1,2,3]
b=['a','b','c']
my_zip(a,b)
```

```python
import numpy as np
x=np.arange(-10,10,1)
#plt.figure(figsize=(5,10)) #그래프의 사이즈
for i in [0.5, 0, 2, 8]: #기울기
    plt.plot(x,i*x, label=f'y={i}x')
plt.xticks(np.arange(min(x), max(x)+1, 1)) #x축의 눈금 범위
plt.grid()
plt.legend(fontsize=14, loc=4)
plt.show()
```

```python
x=np.arange(-10,10) #-10~9
b=np.arange(-5,5,2)
print(b)
for i in b:
    plt.plot(x,x+i, label=f'y=x+{i}')
plt.grid()
plt.legend(fontsize=13)
plt.show()
```

```python
x=np.arange(-10,11)
y=np.zeros(len(x)) #갯수만큼 0을 채움
print(y)
for i in range(len(x)):
    if x[i]>=0:
        y[i]=x[i]
print(y)
plt.plot(x,y)
plt.grid()
plt.show()
#x가 0보다 크면 그대로 출력, x가 음수면 y=0
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-1,7,0.1)
y=5*x-x**2
#print(x)
#print(y)
plt.plot(x,y)
plt.grid()
plt.axhline(y=0, c='r')
plt.axvline(x=0, c='r')
plt.show()
#y=0이 되는 x값은? 0,5
```

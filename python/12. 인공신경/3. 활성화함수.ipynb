```python
import numpy as np
import matplotlib.pyplot as plt
def step_function(x):
    return np.array(x > 0, dtype=np.int64) # 계단함수 (0보다 크면 1, 작으면 0)
X = np.arange(-5.0, 5.0, 0.1)
print(X)
y = step_function(X)
plt.plot(X, y)
plt.ylim(-0.1, 1.1) # y축의 범위 지정
plt.show()
```

```python
# 시그모이드 함수: S자 모양의 함수
import numpy as np
import matplotlib.pyplot as plt
#np.exp() y=e의 x승
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) # 0.0~1.0으로 출력됨
X = np.arange(-5.0, 5.0, 0.1)
print(X)
y = sigmoid(X)
plt.plot(X, y)
plt.ylim(-0.1, 1.1)
plt.show()
#0,1이 아닌 연속적인 실수값을 출력함
```

```python
# Hyperbolic Tangent(tanh)
# 확장된 시그모이드 함수
# 시그모이드는 0~1 사이 출력
# tanh는 -1 ~ 1
# 시그모이드보다 출력범위가 더 넓고 경사면이 더 가파르기 때문에
# 더 빠르게 수렴하여 학습이 이루어질 수 있음
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(-np.pi, np.pi, 12)
y = np.tanh(X)
print(X)
print(y)
plt.plot(X, y, c = 'r', marker = "o")
plt.show()
```

```python
# 입력이 0보다 크면 그대로 출력, 0이하이면 0을 출력
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0, x)
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
# x>0 => 그대로 출력, x<=0 => 0
```

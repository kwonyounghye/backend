```python
import pandas as pd
df=pd.read_csv('c:/data/guerry/data.csv')
df
```

```python
df.plot()
```

```python
# ê²°ì¸¡ê°’ 1ê°œê°€ ì œê±°ë¨, inplace=True ì‹¤í–‰ í›„ ê²°ê³¼ê°’ì„ í˜„ìž¬ ë³€ìˆ˜ì— ë‹¤ì‹œ ì €ìž¥(df ì›ë³¸ êµì²´)
df.dropna(inplace=True)
df.shape
```

```python
# ë³€ìˆ˜ 4ê°œë§Œ ì„ íƒ
df2 = df[['Crime_prop', 'Literacy', 'Wealth', 'Distance']] # ì²«ë²ˆì§¸ ì¢…ì†ë³€ìˆ˜, ë‚˜ë¨¸ì§€ ë…ë¦½ë³€ìˆ˜
```

```python
import statsmodels.formula.api as smf
# Ordinary Least Squares(OLS) ë°©ì‹
# ìž”ì°¨(Residual) : ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì˜ ì°¨
# ìž”ì°¨ì˜ ì œê³±ì˜ í•©ì„ ìµœì†Œë¡œ í•˜ëŠ” ë°©ì‹ìœ¼ë¡œ íšŒê·€ì„ ì„ ê·¸ë¦¬ëŠ” ë°©ì‹
model = smf.ols(formula='Crime_prop ~ Literacy + Wealth + Distance', data=df2).fit() # fit íŒ¨í„´ ê²€ìƒ‰, í•™ìŠµ
model.summary()
# R-squared: ì„¤ëª…ë ¥
```

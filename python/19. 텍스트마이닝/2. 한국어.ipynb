```python
from konlpy.tag import Hannanum
han=Hannanum() # 한나눔 패키지
txt=""" 원/달러 환율이 3년 5개월 만에 최고치로 마감하고
위안화 환율이 11년 만에 달러당 7위안을 넘었다.
원/엔 재정환율 역시 100엔당 30원 가까이 뛰었다."""
#형태소 분석
print(han.morphs(txt)) # 형태소 분석
```

```python
print(han.nouns(txt)) #명사 추출
```

```python
#형태소와 품사
print(han.pos(txt))
```

```python
print(han.tagset) #품사 목록
```

```python
# 꼬꼬마 패키지 - IDS(Intelligent Data Systems) 서울대학교 연구실 개발
# http://kkma.snu.ac.kr/
from konlpy.tag import Kkma
kkm=Kkma()
print(kkm.morphs(txt)) #형태소 분석
```

```python
print(kkm.nouns(txt)) # 명사만 추출
```

```python
print(kkm.pos(txt))
```

```python
print(kkm.tagset) #품사 목록
```

```python
# 트위터 패키지 - 트위터에서 개발, 2017년 이후 오픈코리안텍스트로 이름이 바뀜
# https://github.com/twitter
from konlpy.tag import Okt
okt=Okt()
print(okt.morphs(txt)) #형태소 분석
```

```python
print(okt.nouns(txt))
```

```python
print(okt.pos(txt))
```

```python
print(okt.tagset)
```

```python
#어구 추출
print(okt.phrases(txt))
```

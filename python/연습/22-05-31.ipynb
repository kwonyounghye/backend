```python
#지도학습: 답 알려주고 학습 시킴
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) # 가장 가까운 이웃의 수 3
knn.fit(X, y)
# 1.1의 분류는?
print(knn.predict([[1.1]])) # 0으로 분류됨
# 0.9는 0일 확률 66%, 1일 확률 33% => 0으로 분류됨
print(knn.predict_proba([[0.9]])) # 0일 확률, 1일 확률
print(knn.predict([[0.9]]))
print(knn.predict([[1.9]]))
```

```python
#비지도학습
samples = [[0., 0., 0.], [0., .5, 0.], [1., 1., .2]]
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(n_neighbors=1)
nn.fit(samples)
# 새로운 데이터 1,1,1에 가장 가까운 데이터는? [1, 1, 0.2]
print(nn.kneighbors([[1, 1, 1]]))
# 세번째가 가장 가깝고 거리는 0.8
```

```python
# 최근접 이웃 회귀모형
from sklearn.neighbors import KNeighborsRegressor
X = [[40], [45], [60], [70]] # 평방미터
y = [20, 22, 30, 35] # 월세
neigh = KNeighborsRegressor(n_neighbors=2) # 회귀모형
neigh.fit(X, y)
# 65 평방미터일 때의 월세는?
print(neigh.predict([[65]]))
print(neigh.predict(X))
```

```python
from sklearn.model_selection import train_test_split
#데이터셋 분리
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 3개의 이웃
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# 데이터 학습
knn.fit(X_train, y_train)
```

```python
print(" 검증용 데이터셋의 예측값: {0}".format(knn.predict(X_test)))
```

```python
print(" 검증용 데이터셋의 예측 정확도: {0:.1f}%".format(knn.score(X_test, y_test) * 100))
```

```python
import matplotlib.pyplot as plt
import numpy as np
X=[10,25,47,68,89]
y=[20,40,55,65,75]
plt.figure(figsize=(5,5))
plt.scatter(X,y)
mypoint=(80,5)
# plt.scatter(x좌표, y좌표, size, color)
plt.scatter(mypoint[0],mypoint[1],s=100,c='r')
plt.show()
```

```python
# a,b의 유클리드 거리 구하기
from math import sqrt
def distance(p, m):
    items=[]
    for i in range(len(p-1)):
        distance = sqrt((m[0] - p[i][0])**2 + (m[1] - p[i][1])**2)
        items.append(distance)
    return items
points=np.array(list(zip(X,y)))
r=distance( points, mypoint )
print(r)
m=min(r)
i=np.argmin(r)
print('가장 가까운 포인트의 인덱스: ', i)
print('가장 가까운 샘플: ', points[i])
```

```python
#사이파이 패키지에 내장된 유클리드(기하학) 거리 함수
from scipy.spatial import distance
results=[]
for point in points:
    dist = distance.euclidean(point, mypoint)
    results.append(dist)
results
```

```python
#X,y의 맨해튼 거리 구하기
def distance(p, m):
    items=[]
    for i in range(len(p-1)):
        distance = abs(m[0] - p[i][0]) + abs(m[1] - p[i][1])
        items.append(distance)
    return items
r=distance( points, mypoint )
print(r)
m=min(r)
i=np.argmin(r)
print(' 가장 가까운 포인트의 인덱스:',i)
print(' 가장 가까운 샘플: ',points[i])
```

```python
#사이파이 패키지에 내장된 맨해튼 거리를 구하는 함수
from scipy.spatial import distance
results=[]
for point in points:
    # cityblock: 맨해튼 거리 계산하는 함수
    dist = distance.cityblock(point, mypoint)
    results.append(dist)
results
```

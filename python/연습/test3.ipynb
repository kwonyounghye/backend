```python
html='''
<html><body>
<h1>hello python</h1>
<p>웹페이지 분석</p>
<p>웹 스크레이핑</p>
</body></html>
'''
# BeautifulSoup(html소스, 분석기)
bs=BeautifulSoup(html, 'html.parser') 
h1=bs.html.body.h1 # h1 태그
p1=bs.html.body.p  # p 태그
p2=p1.next_sibling.next_sibling  #형제 노드
print(h1.text) # 태그.text 태그 내부의 텍스트
print(p1.text)
print(p2.text)
```

```python
html='''
<html><body>
<ul>
<li><a href='http://naver.com'>naver</a></li>
<li><a href='http://daum.net'>daum</a></li>
<li><a href='http://nate.com'>nate</a></li>
<li><a href='http://google.com'>google</a></li>
<li><a href='http://yahoo.com'>yahoo</a></li>
</ul>
</body></html>
'''
bs=BeautifulSoup(html, 'html.parser')
links=bs.find_all('a') #모든 a 태그를 리스트로 저장 
print(links)
for a in links:
    href=a['href'] # 태그의 href 속성값 
    text=a.text # 태그 내부의 텍스트 
    print(text,':',href)
```

```python
category=['20040000','33580000']
for cate in category:
    html=urlopen('http://browse.auction.co.kr/list?category='+cate)
    bs=BeautifulSoup(html.read(), 'html.parser')
    tags=bs.find_all('img') #모든 img 태그 수집 
    for img in tags:
        # 이미지 태그의 src 속성, https가 없으면
        if 'https' not in img['src']: 
            src='https:'+img['src'] #추가
        print(src)
        name=src.split('/')[-1] # /로 나누고 마지막 값을 저장 
        try:
            urllib.request.urlretrieve(src,'c:/data/images/'+name) #저장
        except:
            pass
```

```python
import urllib.request as req 
url='http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
res=req.urlopen(url) #url 접속, html 다운로드
bs=BeautifulSoup(res, 'html.parser')
title=bs.find('title').text # title 태그 내부의 텍스트
wf=bs.find('wf').text 
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]
items=bs.find_all('data') #모든 data 태그 리스트
for item in items:
    times=item.find_all('tmef')
    wfs=item.find_all('wf')
    tmns=item.find_all('tmn')
    tmxs=item.find_all('tmx')
    rnsts=item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)

#같은 인덱스끼리 결합
result=zip(result1,result2,result3,result4,result5)
rows=list(result)
rows
```

```python
import pandas as pd 
df=pd.DataFrame(rows, columns=['일시','날씨','최저','최고','강수확률'])
df.head()
```

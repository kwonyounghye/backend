```python
#매개변수를 추가하여 인터넷 리소스를 요청하는 방법
import urllib.request
import urllib.parse
api = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# url에 특수문자, 한글이 포함된 경우 URL 인코딩이 필요함
# 지역번호 : 전국 108, 서울/경기 109, 강원 105, 충북 131,
# 충남 133, 전북 146, 전남 156, 경북 143, 경남 159, 제주 184
values = { 'stnId': '108' } #전국옵션
#values = { 'stnId': '109' } # 서울/경기
#
params = urllib.parse.urlencode(values) # url에 접속하여 읽음
# 요청 전용 URL 생성
url = api + "?" + params
print("url=", url)
# 다운로드
data = urllib.request.urlopen(url).read() # url에 접속하여 읽음
text = data.decode("utf-8") # utf-8 형식으로 변환
print(text)
```

```python
# 라이브러리 로딩
from bs4 import BeautifulSoup
# 분석하고 싶은 HTML
html = """
<html><body>
<h1>Hello Python</h1>
<p>웹 페이지 분석</p>
<p>웹 스크레이핑</p>
</body></html>
"""
# HTML 분석, html.parser 분석기 사용
# html.parser : 기본 파서
# lxml 등의 외부 파서를 사용할 수 있음.
# BeautifulSoup(html소스, 분석기)
bs = BeautifulSoup(html, 'html.parser')
# 원하는 부분 추출하기
h1 = bs.html.body.h1
# 1개의 p 태그 선택
p1 = bs.html.body.p
# sibling 형제 노드(같은 레벨의 노드)
# previous_sibling - 같은 레벨의 이전 요소
# next_sibling - 같은 레벨의 다음 요소
#첫번째 p 태그의 </p> 뒤의 공백문자, 그 뒤의 <p>태그 내용이 선택됨
p2 = p1.next_sibling.next_sibling # 형제 노드 
# 요소의 글자 출력하기
print("h1 = " + h1.text) # 태그.TEXT 태그 내부의 텍스트
print("p = " + p1.text)
print("p = " + p2.text)
items=bs.find_all('p')#모든 p 태그를 찾음
```

```python
from bs4 import BeautifulSoup
import urllib.request as req
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# urlopen()으로 데이터 가져오기
res = req.urlopen(url)
# BeautifulSoup으로 분석
bs = BeautifulSoup(res, "html.parser")
# 원하는 데이터 추출
# title 태그
title = bs.find("title").text
# wf 태그
wf = bs.find("wf").text
print(title)
print(wf)
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]
items=bs.find_all('data') #data 태그들을 리스트로 저장
#print(items)
for item in items:
    times=item.find_all('tmef')
    wfs=item.find_all('wf')
    tmns=item.find_all('tmn')
    tmxs=item.find_all('tmx')
    rnsts=item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)
print(result1[:5])
print(result2[:5])
#리스트의 같은 인덱스끼리 저장
result=zip(result1,result2,result3,result4,result5)
rows=list(result)
rows
```

```python
import pandas as pd
df=pd.DataFrame(rows,
columns=['일시','날씨','최저','최고','강수확률'])
df.head()
```

```python
from bs4 import BeautifulSoup
# 분석 대상 HTML
html = """
<html><body>
<div id="main">
<h1>도서 목록</h1>
<ul class="items">
<li>빅데이터</li>
<li>HTML</li>
<li>Python</li>
</ul>
</div>
</body></html>
"""
# HTML 분석
bs = BeautifulSoup(html, 'html.parser')
# 필요한 부분을 CSS로 추출
# 타이틀 부분 추출
# id가 main인 div 태그의 자손 태그 h1 선택
# 태그 1개 선택, div 중에서 id가 main,하위에 h1
h1 = bs.select_one("div#main > h1").text
print("h1 =", h1)
# 목록 추출
# id가 main인 div 태그 => class가 items인 ul 태그 => li 태그들을 선택
li_list = bs.select("div#main > ul.items > li")
# 반복 처리
for li in li_list:
    print("li =", li.text)
```

```python
#옥션에서 상품 분류 코드 가져오기
from bs4 import BeautifulSoup
import urllib.request as req
import urllib.request
from urllib.request import urlopen
#상품분류코드를 저장할 리스트
codes=[]
for num in range(51,53):
    url = "http://www.auction.co.kr/category/category{0}.html".format(num)
    print(url)
    res = req.urlopen(url)
    # HTML 분석하기
    bs = BeautifulSoup(res, "html.parser", from_encoding='ms949')
    # 원하는 데이터 추출(카테고리 전체보기를 클릭한 후 개발자 도구에서 패턴 찾기)
    category = bs.select(\
    "http://browse.auction.co.kr/m/list?category=51190000")
    for cate in category:
        name=cate.text
    href=cate.attrs["href"].split("=")[1]
    codes.append(href)
    print("상품분류명:"+name)
    print("분류코드:"+href+"\n")
#category=["33580000","44040000","64050000"]
#데이터를 수집할 url에 접속
for cate in codes:
    html = urlopen("http://browse.auction.co.kr/list?category="+cate)
    # HTML 분석
    bs = BeautifulSoup(html.read(), 'html.parser')
    # find_all() 함수로 a 태그를 모두 추출
    tags = bs.find_all("img")
    # 링크 목록 출력
    for img in tags:
        #a 태그의 속성 중에서 href인 값
        if 'http:' not in img['src']:
            src = 'http:'+img['src']
        print(src)
        # url이 가리키는 주소에 접근해서 해당 자원을 로컬 컴퓨터에 저장하는 코드
        name=src.rsplit('/', 1)[-1]
        urllib.request.urlretrieve(src, "d:/data/images/"+name)
```

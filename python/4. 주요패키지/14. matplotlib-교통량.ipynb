```python
import pandas as pd
df=pd.read_csv("c:/data/traffic/data.csv",encoding='ms949')
df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
# 한글 처리를 위해 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
# 막대 그래프
# R용 시각화 패키지 ggplot의 스타일 적용
plt.style.use('ggplot')
months = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
months_index=range(1,len(months)+1)
result=[]
for i in range(12):
                            #str+int=(x)=>str으로 바꾸기
    result.append(df['2021년'+str(i+1)+'월'].sum()/10000000)
print(result)
```

```python
# figure (그림)객체를 만든다.
fig=plt.figure()
# figure 객체 하위에 그래프를 추가
# 1행1열 1개의 하위 그래프
ax1=fig.add_subplot(1,1,1)
# 막대 그래프 bar (x축, y축, 가운데 정렬, 막대 색상)
ax1.bar(months_index, result, align='center', color='darkblue')
# x축 눈금 위치
ax1.xaxis.set_ticks_position('bottom')
# y축 눈금 위치
ax1.yaxis.set_ticks_position('left')
# xticks (수치 데이터, 대체 텍스트) 숫자 대신 문자열로 출력됨
plt.xticks(months_index, months, rotation=0)
# 눈금 라벨 사이즈
plt.tick_params(labelsize=14)
# 타이틀
plt.xlabel('월', fontsize=20)
plt.ylabel('합계', fontsize=20)
plt.title('통계', fontsize=25)
# 그래프를 이미지로 저장, dpi해상도, tight 그림을 둘러싼 여백 제거
plt.savefig('c:/data/images/bar_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```

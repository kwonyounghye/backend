```python
# 산점도
plt.scatter(X['sepal length (cm)'],X['sepal width (cm)'], label='sepal')
plt.scatter(X['petal length (cm)'],X['petal width (cm)'], label='petal')
plt.legend()
plt.show()
```

```python
from sklearn import datasets
iris=datasets.load_iris() 
iris.target
```

```python
import numpy as np
#그래프 사이즈
plt.figure(figsize=(14,7))
#색상 배열
colormap=np.array(['red','lime','black'])
#꽃받침 기준 그래프
plt.subplot(1,2,1) # 1행2열 중 1열의 그래프
plt.scatter(X['sepal length (cm)'],X['sepal width (cm)'],c=colormap[iris.target],s=40) # s점의 크기
plt.title('꽃받침')
# 꽃잎 기준 그래프
#plt.figure(figsize=(14,7))
plt.subplot(1,2,2) # 1행2열 중 2열의 그래프
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=colormap[iris.target], s=40)
plt.title('꽃잎')
```

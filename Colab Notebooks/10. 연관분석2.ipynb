```python
import re
df['clean_review']=df['review'].str.replace('[^a-zA-Z]', ' ')
df['clean_review']=df['clean_review'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
df['clean_review']=df['clean_review'].apply(lambda x:x.lower())

X=df['clean_review']
```

```python
import nltk
#nltk.download('punkt')
#cnt=0
for i,document in enumerate(X.values[:1000]):
  #cnt+=1
  clean_words = []
  words=document.split()
  #print(words)
  for word in words:
    if word not in nltk.corpus.stopwords.words('english'): #불용어 제거
      clean_words.append(word)  
  document = ' '.join(clean_words)
  #print(document) #stale uninspir
  X[i] = document      
#  if cnt>10: 
#    break 

#X[:10]
```

```python
items2=[]
for row in X[:1000]:
    arr=row.split()
    line=[]
    for a in arr:
        line.append(a)
    items2.append(line)

items2
```

```python
X.values[0]
```

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
te = TransactionEncoder()
t_arr = te.fit(items2).transform(items2)
print(te.columns_)
print(t_arr)
df2 = pd.DataFrame(t_arr, columns = te.columns_)
df2
```

```python
from mlxtend.frequent_patterns import apriori
frequent = apriori(df2, min_support=0.2, use_colnames=True)
frequent
```

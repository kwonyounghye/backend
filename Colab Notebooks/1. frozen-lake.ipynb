```python
import gym
# 환경(environment) : 환경 문제를 구성하는 모든 규칙
# 에이전트(agent) : 에이전트 주어진 문제 상황에서 행동하는 주체
# 정책(policy) : 정책 에이전트가 판단하는 방식
# 상태(state) : 상태 현재 시점에서 상황이 어떤지 나타내는 값의 집합
# 행동(action) : 행동 현재 상황에서 취할지 있는 동작
# 보상(reward) : 보상 에어전트가 어떤 행동을 했을 때 따라오는 이득
# 에피소드(episode) : 에피소드 초기 상태부터 종료 상태까지 에어전트가 거친 행동의 과정
# 환경 생성하기(미로 탐색 게임)
env = gym.make('FrozenLake-v0')
```

```python
# 게임 시작
# 20 , 1 100 0.78 게임 회 게임 회당 턴 수행해서 얻은 평균 보상값이 이상이 되면 게임이 종료됨
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        # 필드 그리기
        env.render()
        # ( ) 무작위로 행동을 선택 상하좌우
        action = env.action_space.sample()
        # 1턴 진행
        # observation : 이동 후의 위치
        # reward : ( -1 , 보상 함정에 빠지면 점 목표지점에 도달하면1 ) 점
        # done : 게임 종료 여부
        # info : 디버깅 등에 필요한 변수
        observation, reward, done, info = env.step(action)
        print('observation:',observation)
        print('reward:',reward)
        print('done:',done)
        print('info:',info)
        print('action:', action)
        # 종료
        if done:
            print("Episode finished after {} timesteps".formt(t+1))
            break
```

```python
# 환경변수 값의 경우의 수
print(env.observation_space.n)
# 행동변수 값의 경우의 수
print(env.action_space.n)
```

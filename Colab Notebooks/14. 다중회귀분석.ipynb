```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X=df[["period", "votes", "score", "runtime"]]
y=df["gross"]
model.fit(X,y)
```

```python
import statsmodels.formula.api as smf
import numpy as np
# Ordinary Least Squares(OLS) 방식
# 잔차(Residual) : 실제값과 예측값의 차
# 잔차의 제곱의 합을 최소로 하는 방식으로 회귀선을 그리는 방식
# coef : 기울기, intercept : 절편, std err : 표준오차, r-squared : 결정계수
model = smf.ols(formula='gross ~ period + score + votes + runtime', data=df).fit()
model.summary()
```
